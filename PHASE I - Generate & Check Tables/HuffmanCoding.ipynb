# Formal Process using Heap

```python
from heapq import heappush, heappop, heapify
from collections import defaultdict
```

```python
# text = simpleList
text=simpleList
```

```python
freq_lib = defaultdict(int)    # generate a default library
for ch in text:                # count each letter and record into the frequency library 
    freq_lib[ch] += 1
    
print(freq_lib)
print(len(freq_lib))
```

```python
heap = [[fq, [sym, ""]] for sym, fq in freq_lib.items()]  # '' is for entering the huffman code later
print(heap)
```

```python
heapify(heap) # transform the list into a heap tree structure
print(heap)
```

```python
while len(heap) > 1:
    print(len(heap))
    right = heappop(heap)  # heappop - Pop and return the smallest item from the heap
    print('right = ', right)
    left = heappop(heap)
    print('left = ', left)

    
    print("right: =",right[1:])
    print("left: =",left[1:])
    for pair in right[1:]:  
        pair[1] = '0' + pair[1]   # add zero to all the right note
    for pair in left[1:]:  
        pair[1] = '1' + pair[1]   # add one to all the left note
    
    print("Edited left: ",left)
    print("Edited right: ",right)
    
    heappush(heap, [right[0] + left[0]] + right[1:] + left[1:])  # add values onto the heap. Eg. h = []; heappush(h, (5, 'write code')) --> h = [(5, 'write code')]
```

```python
df=pd.DataFrame(columns=["Huffman Code","Frequency"])
df.index.name="Values"
totalList=right[1:]+left[1:]
for i in totalList:
    df.loc[i[0]]=["_"+(i[1])+"_",freq_lib[i[0]]]

display(df.head())
```

```python
df.to_csv("YAH.bod", encoding='utf-8', index = True)
```
